# **2. Read alignment**


---

```python
#Write your commands here. Insert more cells if necessary
!gunzip ~/scratch/Project2/ReadAlignment/hg38.fa.gz
```

```python
#Run "ls -lh" to show what additional files were created in addition to your original fasta files.
!ls -lh ~/scratch/Project2/ReadAlignment/
```

```python
!bwa index -a bwtsw ~/scratch/Project2/ReadAlignment/hg38.fa
```

```python


!bwa mem -t 24 ~/scratch/Project2/ReadAlignment/hg38.fa ~/scratch/Project/data/Trimming/SRR099988_1.Trimmed.fastq.gz ~/scratch/Project/data/Trimming/SRR099988_2.Trimmed.fastq.gz> ~/scratch/Project2/ReadAlignment/SRR099988.sam
```

```python
'''
for index, I used bwtsw for algorithm as it works well with whole human genome data
bwa: burrows-Wheeler Aligner software packages to map non-reference genome
mem: maximal exact match
-t 24: -t argument calls the number of CPU cores for the alinment process. In this command, we use 24 CPU cores for the alignment.

'''
```

```python
!minimap2 -d ~/scratch/Project2/ReadAlignment/hg38.mmi ~/scratch/Project2/ReadAlignment/hg38.fa
```

```python
'''
-d: this option will save the index of target fastsa file as a new file (in our case, that'll be an .mmi file)
the first path is the path to an output file, the next path is the one to an input file
'''
```

```python
!minimap2 -c -a ~/scratch/Project2/ReadAlignment/hg38.mmi ~/scratch/Project/data/Trimming/SRR099988_1.Trimmed.fastq.gz ~/scratch/Project/data/Trimming/SRR099988_2.Trimmed.fastq.gz> ~/scratch/Project2/ReadAlignment/SRR099988_minimap.sam  
```

```python
'''
-c: an input/output option that generates Compact Idiosyncratic Gapped Alignment Report (CIGAR).
-a: an input/output option that creates CIGAR and output alignments in the SAM format.
the first path is minimap indexed hg38 reference genome
the second and first paths are two outputs for the alignment
the path after ">" refers to an output
'''
```

```python
!wc -l ~/scratch/Project2/ReadAlignment/SRR099988_minimap.sam

'this file has 176131465 total lines'
```

```python
!samtools view -c ~/scratch/Project2/ReadAlignment/SRR099988_minimap.sam
'this file has 176131008 lines for alignment records'
```
