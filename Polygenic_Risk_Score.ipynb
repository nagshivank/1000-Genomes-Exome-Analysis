```python
#Put all the code and comments here.
#Show that the variant IDs are also overlapping.
#merge all 22 vcf files
!bcftools concat -o ~/scratch/Project4/1000genome_vcf_files/AMR_combined.vcf.gz ~/scratch/Project4/1000genome_vcf_files/phase3.chr{1..22}.GRCh38.GT.crossmap.AMR_samples.vcf.gz
#download PGS data
!wget https://ftp.ebi.ac.uk/pub/databases/spot/pgs/scores/PGS000894/ScoringFiles/Harmonized/PGS000894_hmPOS_GRCh38.txt.gz -P ~/scratch/Project4/
#extract chr number and positions from PSG file
!zcat ~/scratch/Project4/PGS000894_hmPOS_GRCh38.txt.gz | grep -v "#" | awk 'NR>1{ print "chr"$1"\t"$2 }' > ~/scratch/Project4/PGS000894_hmPOS_GRCh38_trimmed.txt
!zcat ~/scratch/Project4/1000genome_vcf_files/AMR_combined.vcf.gz | grep -v "^#" |awk '{ print $1"\t"$2 }' > ~/scratch/Project4/1000genome_vcf_files/AMR_combined_trimmed.vcf

vcf_variants = set()
with open('/home/hice1/ynitta6/scratch/Project4/1000genome_vcf_files/AMR_combined_trimmed.vcf', 'r') as vcf_file:
    for line in vcf_file:
        variant_entry = tuple(line.strip().split()[:2])
        vcf_variants.add(variant_entry)

# Read PGS file (assuming it's a text file with a column containing variant IDs)
pgs_variants = set()
with open('/home/hice1/ynitta6/scratch/Project4/PGS000894_hmPOS_GRCh38_trimmed.txt', 'r') as pgs_file:
    # Assuming the variant IDs are in the first column
    for line in pgs_file:
        variant_entry = tuple(line.strip().split()[:2])
        pgs_variants.add(variant_entry)

# Count duplicates
duplicate_variants = vcf_variants.intersection(pgs_variants)
num_duplicates = len(duplicate_variants)

print(f'Number of duplicate variants: {num_duplicates}')
```

```python
import pandas as pd
from scipy.stats import zscore
import seaborn as sns
from matplotlib.legend_handler import HandlerTuple
```

```python
scores = pd.read_csv('~/scratch/Project4/scores/PGS000894.PRS.sscore', sep="\t")
scores.head()
```

```python
pop_map = pd.read_csv("~/biol6150/Data/1000GenomesMap.tsv", sep="\t", header=None, names=["#IID", "POP"])
full_scores = pd.merge(scores, pop_map, on='#IID', how='left')
```

```python
full_scores['ScaledPRS'] = zscore(full_scores['SCORE1_AVG'])
full_scores.head()
```

```python
#Plot some boxplots by Population.
ax = sns.boxplot(data=full_scores, x="POP", y="ScaledPRS", boxprops={'alpha': 0.4})
```
