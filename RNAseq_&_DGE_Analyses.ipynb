The RNA-seq files are from the experiment which aims to determine the up- and down-regulated genes involved in angiogenesis and vasculogenesis, which helps to develop a potency assay for CD34+ cell-based therapy.

Total RNA was purified from ProtheraCytes® collected from 7 Frozen healthy donor and 8 Acute myocardiac infarcted patients using RNAeasy plus minikit (QIAGEN,Courtabeuf, France). Briefly, sample quality was assessed using a DNA High sensitivity chip (Agilent Technologies). Single end 50 reads barcoded RNA-Seq sequencing library was performed on an Illumina HiSeq4000 with the TruSeq SBS v3 chemistry.


According to a paper from Tirdea et al., LGALS2, and APOA5 gene are two the genes that increase the risk for AMI. 

Research by Wang & Xian found that NR4A3 gene is highly associated with the mechanism of AMI.



### Thus, we assign LGALS2, APOA5, and NR4A3 genes as our genes of interest.

Tirdea, C., Hostiuc, S., Moldovan, H., & Scafa-Udriste, A. (2022). Identification of Risk Genes Associated with Myocardial Infarction-Big Data Analysis and Literature Review. International journal of molecular sciences, 23(23), 15008. https://doi.org/10.3390/ijms232315008


Wang, Y., & Xian, H. (2022). Identifying Genes Related to Acute Myocardial Infarction Based on Network Control Capability. Genes, 13(7), 1238. https://doi.org/10.3390/genes13071238

```python
import os
import shutil

quant_dir = "/home/hice1/ynitta6/scratch/Project5and6/RNAseq_analysis/Quant"

output_dir = "/home/hice1/ynitta6/scratch/Project5and6/RNAseq_analysis/Quant_sf_files"

for srr_dir in os.listdir(quant_dir):
    srr_dir_path = os.path.join(quant_dir, srr_dir)

    if os.path.isdir(srr_dir_path):
        srr_id = srr_dir.split("_")[0]

        quant_sf_file = os.path.join(srr_dir_path, "quant.sf")

        if os.path.exists(quant_sf_file):
            output_quant_sf_file = os.path.join(output_dir, f"{srr_id}_quant.sf")

            shutil.move(quant_sf_file, output_quant_sf_file)
```

```python
!ls -lh ~/scratch/Project5and6/RNAseq_analysis/Quant_sf_files
```

```python
!mkdir -p "/home/hice1/ynitta6/scratch/Project5and6/RNAseq_analysis/Quant_sf_files/HC"
!mkdir -p "/home/hice1/ynitta6/scratch/Project5and6/RNAseq_analysis/Quant_sf_files/Disease"
input_dir = "/home/hice1/ynitta6/scratch/Project5and6/RNAseq_analysis/Quant_sf_files"
output_dir1 = "/home/hice1/ynitta6/scratch/Project5and6/RNAseq_analysis/Quant_sf_files/HC"
output_dir2 = "/home/hice1/ynitta6/scratch/Project5and6/RNAseq_analysis/Quant_sf_files/Disease"

HC_srr_ids = ["SRR22711834", "SRR22711836", "SRR22711839", "SRR22711842", "SRR22711845", "SRR22711848", "SRR22711851"]

Disease_srr_ids = ["SRR22711854", "SRR22711857", "SRR22711861", "SRR22711864", "SRR22711867", "SRR22711870", "SRR22711873", "SRR22711876"]

for sf_file in os.listdir(input_dir):
    srr_id = sf_file.split("_")[0]
    if srr_id in HC_srr_ids:
        input_file_path = os.path.join(input_dir, sf_file)

        output_file_path = os.path.join(output_dir1, sf_file)

        shutil.copy(input_file_path, output_file_path) 
        
    elif srr_id in Disease_srr_ids:
        input_file_path = os.path.join(input_dir, sf_file)

        output_file_path = os.path.join(output_dir2, sf_file)

        shutil.copy(input_file_path, output_file_path) 
        
```
