```python
'''
A pile-up format refers to a format that summarizes the information of aligned sequences of DNA, RNA, or protein. 
It displays multiple sequences aligned to a reference sequence. This format also provides a clear 
view of the variations, insertions, deletions, and single-nucleotide polymorphisms (SNPs) present 
in the aligned sequences.
''''''
```

```python
'''
The columns include-
1.Chromosome/Reference Name: This serves as the identifier for the reference or chromosome sequence which was used to align the reads
2.Base Position: This column contains integer values pointing to the specific base positions (indexed from 1) in the reference sequence where the pileup data is provided
3.Reference Base: This column includes the nucleotide base (A, T, G, C) or ’N’ for an unknown or undetermined position in the reference sequence
4.Read Count: This column contains integer values for the number of reads aligned to that base position
5.Read Results/Bases: This column contains a string containing a string to depict the bases at this position in the reads relative to the reference sequence. It has symbols for the base (A,T,G,C) and various symbols (such as ,+.^$) to show matches, mismatches and indels relative to the sequence.
6.Base Quality: This column contains an ASCII-33 string to show the Phred quality scores of the reads corresponding to the read bases.
'''
```

```python
'''
We used 'mpileup2snp' command to calls indels from a mpileup file. 
--min-reads2 10 as a minimum supporting reads to call variants from our trimmed fasta file that has 
a read size of 60. 

Our trimmed fasta file has a read size of 60 -> We guessed 6~12 (10 in this time) 
would be an appropriate number for the threshold.
'''
```
